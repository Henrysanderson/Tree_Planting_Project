# Tree Planting Project Analysis


```python
# Importing the needed packages
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
#Loading the dataset
df = pd.read_csv('/Desktop/BETA/DATA_SCIENCE/Analysis/Tree Project/tree_planting_project_dataset.csv')
```

```python
# Viewing summary information about the dataset
df.info()
```

```python
# Changing the date to the correct datetime format
df['Planting_Date'] = pd.to_datetime(df['Planting_Date'])
df['Monitoring_Date'] = pd.to_datetime(df['Monitoring_Date'])
df.info()
```

```python
#  Viewing the first 10 rows of the dataset
df.head(10)
```

```python
# Checking the number of null values
df.isnull().sum()
```

```python
# Filling up the missing height values
# Fill with the median
df['Current_Height (cm)'] = df['Current_Height (cm)'].fillna(df['Current_Height (cm)'].median())
df.isnull().sum()
```

```python
# Droping duplicates
df.drop_duplicates()
```

### 1. *Which species has the highest survival rate?*

```python
# Count survived trees by species
survived_by_species = df[df['Survival_Status'] == 'Yes'].groupby('Species')['Survival_Status'].count()
Total_trees = df['Tree_ID'].count()
Survival_Rate = (survived_by_species / Total_trees) * 100
print(round(Survival_Rate, 2))
```

> **Insights:**
Species with higher survival rates may indicate better adaptation to local conditions or more effective planting techniques.

### 2. *How does soil quality affect tree growth?*

```python
Tree_height = df.groupby('Soil_Quality')['Current_Height (cm)'].mean()
print(round(Tree_height, 2))
```

> **Insights:**
Higher soil quality is expected to correlate with better growth metrics (height and diameter). Poor soil quality might limit growth, even if survival is high.

### 3. *What is the relationship between community involvement and biodiversity impact?*

```python
Biodiversity = df.groupby('Community_Engagement_Score')
Biodiversity_Impact = Biodiversity['Species'].nunique()
print(Biodiversity_Impact)
```

> **Insights:**
Higher community involvement might lead to greater biodiversity due to better planting and maintenance practices

### 4. *Is there a significant difference in survival rates across locations?*

```python
# Count survived trees by location
survived_by_species = df[df['Survival_Status'] == 'Yes'].groupby('Location')['Survival_Status'].count()
Total_trees = df['Tree_ID'].count()
Survival_Rate = (survived_by_species / Total_trees) * 100
print(round(Survival_Rate, 2))
```

> **Insights:**
Locations with lower survival rates might need intervention, such as better soil preparation or more watering.
